```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("covid_data.csv")
df = df[['location', 'date', 'total_cases']]
df['date'] = pd.to_datetime(df['date'])
df.groupby(df.date.dt.year)['total_cases'].sum()

```

```python
poland = df[df['location'] == 'Poland']
poland
```

```python
x = poland['date']
y = poland['total_cases']
plt.figure(figsize=(20,10))
plt.plot(x,y)
plt.xlabel('date')
plt.ylabel('total cases in millions')
plt.title('COVID-19 cases in Poland')
plt.show()
```
